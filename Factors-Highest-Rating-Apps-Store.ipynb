```python
#Evaluating the Data Provided in each variable through Error Minimisation, after that obtaining confidence intervals:

#We calculate error minimized mean value as x_bar values.
```

```python
#Error Minimisation For Reviews:
Reviews1 = data['Reviews']

# Define a function to compute the squared errors
def squarerr_Reviews(mu, y):
    return np.square(y - mu)

# Find out the sum of squared errors evaluated at different mu values.

mu_trial = np.linspace(start = np.amin(Reviews1), stop = np.amax(Reviews1), num = 7390)

error_trial = np.empty(7390, dtype = float)
for i in range(7390):
    sumerr_Reviews = np.sum(squarerr_Reviews(mu_trial[i], Reviews1)) 
    error_trial[i] = sumerr_Reviews
```

```python
min_index = np.argmin(error_trial)
best_mu_reviews = mu_trial[min_index]

plt.plot(mu_trial, error_trial)
plt.yscale('log')
plt.xlabel('mu value')
plt.ylabel('log error')
plt.axvline(x=best_mu_reviews, c = "k", ls = "--")


print('Through the graphical method, the x-bar value for Reviews to minimize total squared errors is', best_mu_reviews)
print('Even though the mean is', np.mean(Reviews1))
```

```python
#Next, we calculate the Confidence Interval of each x-variable.
#This is because we do not know which data points are considered outliers as we do not know the true population variance, adding on to not knowing the distribution in real-world data.
# Hence we construct a CI using CLT by just using a good estimate of population variance, say sample variance to compute the standard error

# After constructing the CIs, we will know if there are enough data in each x-variable to have a significant relationship with y-variable: Ratings
```

```python
#Confidence Interval for Reviews:

M = sps.norm.ppf(0.975, loc = 0, scale=1)
SIZE = 7390
REPEAT = 10000

REVIEWS_MEAN = np.mean(data.Reviews)
REVIEWS_VAR = np.var(data.Reviews, ddof = 0)

sample_reviews = np.random.normal(REVIEWS_MEAN, REVIEWS_VAR, (REPEAT,SIZE))
print(sample_reviews.shape)

# Specify axis = 1 to collapse the column  
# Want to get the sample mean of each simulated sample through repeated sampling
sample_reviews_mean = np.mean(sample_reviews, axis = 1) 
sample_reviews_var = np.var(sample_reviews, axis = 1, ddof = 1)
se_xbar_reviews = np.sqrt(sample_reviews_var/SIZE) 
errorwidth_reviews = M * se_xbar_reviews

# Next, count the numeber of the simulated CIs that cover the true population mean
LowerEnd_reviews = sample_reviews_mean - errorwidth_reviews
UpperEnd_reviews = sample_reviews_mean + errorwidth_reviews

coverProb_reviews = np.sum((LowerEnd_reviews < REVIEWS_MEAN) & (UpperEnd_reviews > REVIEWS_MEAN)) / REPEAT
print('In the simulation, the coverage probability for Reviews is', coverProb_reviews)
```
